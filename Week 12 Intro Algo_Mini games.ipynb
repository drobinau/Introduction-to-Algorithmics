#### A Turtle Drawing Machine

```python
#A turtle drawing machine, version 2, created by Robin Wang 
from turtle import *
import time
exception = ['distance', 'towards']

def turtle_move2(action, val, string):
    if action in exception:
        write(eval(string))
        time.sleep(1)
        undo()
    elif val == 0:
        if action=='clear':
            clear()
        else:
            write(eval(action)())
            time.sleep(1)
            undo()
    else:
        eval(string)

def string_converter2(string):
    command = string #you can type in multiple turtle commands, separated by comma. e.g. goto(100,100),forward(100)
    print(command)
    cmd_len = len(command)
    cmd_type = command[0:command.index('(')]
    num = 0
    if cmd_len > 1:
        num_string = command[command.index('(')+1:command.index(')')]
        if num_string != '':
            num = 1
    turtle_move2(cmd_type, num, string)
####
instructions = '''Type in Turtle commands'''
while True:
    typein = textinput('Turtle Drawing Machine', instructions)
    if typein == '':
        continue
    if typein.upper() == 'BYE':
        break
    string_converter2(typein)
bye()
```

#### A control panel 

```python
from tkinter import *
import turtle

global x
global y 

x = 1
y = 1
i = 0

root = Tk()

wn = turtle.Screen()
bob = turtle.Turtle()
bob.shape('turtle')
bob.color('green')
bob.up()

def goUp():
    global x, y
    y=y+5
    bob.goto(x,y)
def goDown():
    global x, y
    y=y-5
    bob.goto(x,y)
def goRight():
    global x, y
    x=x+5
    bob.goto(x,y)
def goLeft():
    global x, y
    x=x-5
    bob.goto(x,y)

topFrame = Frame(root)
topFrame.pack()
middleFrame = Frame(root)
middleFrame.pack()
bottomFrame = Frame(root)
bottomFrame.pack()

button1 = Button(topFrame, text = "Up", fg="red", command=goUp).pack()
button2 = Button(middleFrame, text = "Left", fg="red", command=goLeft).pack(side="left")
button3 = Button(middleFrame, text = "Right", fg="red", command=goRight).pack(side="right")
button4 = Button(bottomFrame, text = "Down", fg="red", command=goDown).pack()

root.mainloop()
```

#### Koch Snowflake Demo

```python
import turtle
def koch(t, order, size):
    if order == 0:
        t.forward(size)
    else:
        for angle in [60, -120, 60, 0]:
           koch(t, order-1, size/3)
           t.left(angle)

t=turtle.Turtle()
t.penup()
t.goto(-100,0)
t.pendown()
t.speed(0)
koch(t,3,300)

turtle.exitonclick()
```

#### An Apple Shooter

```python
import turtle
import math
import random
import time

# --- Setup screen ---
screen = turtle.Screen()
screen.title("Apple Shooter - Angle and Speed Game")
screen.bgcolor("lightblue")
screen.setup(width=800, height=600)

# --- Draw ground ---
ground = turtle.Turtle()
ground.hideturtle()
ground.penup()
ground.goto(-400, -250)
ground.pendown()
ground.color("green")
ground.begin_fill()
for _ in range(2):
    ground.forward(800)
    ground.right(90)
    ground.forward(100)
    ground.right(90)
ground.end_fill()

# --- Score and Timer Setup ---
score = 0
start_time = time.time()
duration = 10 # seconds

display = turtle.Turtle()
display.hideturtle()
display.penup()
display.goto(-380, 260)

def update_display():
    elapsed = int(time.time() - start_time)
    remaining = max(duration - elapsed, 0)
    display.clear()
    display.write(f"Score: {score}    Time Left: {remaining}s", font=("Arial", 16, "bold"))
    if remaining > 0:
        screen.ontimer(update_display, 1000)
    else:
        screen.textinput("Time's up!", f"Final Score: {score}\nDo you like this game?\nLeave your answer below, then press Enter to exit.")
        turtle.bye()

update_display()

# --- Cannon setup ---
cannon = turtle.Turtle()
cannon.shape("square")
cannon.color("black")
cannon.shapesize(stretch_wid=1, stretch_len=3)
cannon.penup()
cannon.goto(-350, -200)
cannon.setheading(45)

# --- Apple setup ---
apple = turtle.Turtle()
apple.shape("circle")
apple.color("red")
apple.penup()
apple_speed = 2

def reset_apple():
    apple.goto(random.randint(-100, 300), 250)

reset_apple()

# --- Projectile setup ---
projectile = turtle.Turtle()
projectile.shape("circle")
projectile.color("orange")
projectile.penup()
projectile.hideturtle()

# --- Game logic ---
def fire(angle, speed):
    global score
    if time.time() - start_time > duration:
        return

    projectile.goto(-350, -200)
    projectile.setheading(angle)
    projectile.showturtle()

    vx = speed * math.cos(math.radians(angle))
    vy = speed * math.sin(math.radians(angle))
    x, y = -350, -200
    g = 9.8
    t = 0

    while y >= -250 and x <= 400:
        t += 0.1
        x = -350 + vx * t
        y = -200 + vy * t - 0.5 * g * t * t
        projectile.goto(x, y)
        screen.update()

        # Collision detection
        if projectile.distance(apple) < 20:
            score += 1
            print("Hit! Score:", score)
            reset_apple()
            break

    projectile.hideturtle()
    screen.ontimer(get_user_input, 500)

# --- User input handler ---
def get_user_input():
    if time.time() - start_time > duration:
        return
    try:
        entry = screen.textinput("Fire!", "Enter angle and speed (e.g. 45,60):")
        if entry:
            angle_str, speed_str = entry.split(",")
            angle = float(angle_str.strip())
            speed = float(speed_str.strip())
            fire(angle, speed)
    except:
        print("Invalid input. Try again.")
        screen.ontimer(get_user_input, 500)

# --- Keyboard binding ---
screen.listen()
screen.onkey(get_user_input, "space")

# --- Apple falling loop ---
def fall_apple():
    if time.time() - start_time > duration:
        return
    y = apple.ycor()
    y -= apple_speed
    if y < -250:
        reset_apple()
    else:
        apple.sety(y)
    screen.ontimer(fall_apple, 50)

fall_apple()
screen.mainloop()
```
